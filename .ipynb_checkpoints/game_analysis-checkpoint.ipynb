```python
%matplotlib notebook
import pandas as pd
import matplotlib as mlp
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('vgsales.csv')
df.head()
```

```python
df=pd.read_csv('vgsales.csv')
df=df[(df['Year']>=2000) & (df['Year']<=2010)]
df=df[['Year','Platform','NA_Sales','EU_Sales','JP_Sales','Other_Sales']]
df.Year=df.Year.astype(int)
df.head()
```

```python
sales_min=min(df['NA_Sales'])
sales_max=max(df['NA_Sales'])
first_year=min(df['Year'])
last_year=max(df['Year'])
df=df.groupby(['Year']).sum()
df
```

```python
years=list(range(first_year,last_year+1))
years_str = [str(i) for i in range(first_year,last_year+1)]
df.plot()
plt.title('Video Games Sales')
plt.xlabel('Years')
plt.ylabel('Sales(in million dollars)')
plt.xlim(first_year,last_year)
plt.xticks(years,years_str)
```
